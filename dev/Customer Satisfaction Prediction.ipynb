```python
# first neural network with keras tutorial
# importing libraries
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.optimizers import Adam

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
```

```python
# load the dataset
df = pd.read_csv(r"C:\Users\zulhu\Desktop\Customer Satisfaction Prediction\feat_eng.csv")
orders_features = df.drop('review_score', axis=1)
orders_labels = df['review_score'].copy()

# split into input (X) and output (y) variables
X = orders_features.values
y = orders_labels.values

# Load test dataset
df = pd.read_csv(r"C:\Users\zulhu\Desktop\Customer Satisfaction Prediction\feat_eng_test.csv")
test_features = df.drop('review_score', axis=1)
test_labels = df['review_score'].copy()

# split into input (X) and output (y) variables
Xt = test_features.values
yt = test_labels.values
```

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X)

norm_X = transformer.transform(X)
norm_Xt = transformer.transform(Xt)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(2048, input_dim = 121, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(1024, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation = 'sigmoid'))

# compile the keras model
adam = Adam(lr=0.001)
model.compile(loss = 'binary_crossentropy', optimizer = adam, metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(norm_X, y, validation_data = (norm_Xt, yt), epochs = 500, batch_size = 1000)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(norm_X, y)
print('Train Accuracy: %.2f' % (accuracy*100))

# evaluate the keras model
_, accuracy = model.evaluate(norm_Xt, yt)
print('Test Accuracy: %.2f' % (accuracy*100))
```

```python
# save model
model.save('model.h5')
```

```python
result = model.predict(norm_X)
#result_argmax = np.around(result)
result_argmax = (result > 0.5).astype(np.int)
result_argmax = result_argmax.flatten()
print(classification_report(y, result_argmax))
```
